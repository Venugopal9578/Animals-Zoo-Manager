# Animal Discovery Analysis 🦁

This notebook provides a basic analysis of the species collected in our `animals.db` database.

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
%matplotlib inline
sns.set_theme(style="whitegrid")
```

## 1. Load Data

```python
conn = sqlite3.connect('animals.db')
df = pd.read_sql_query("SELECT * FROM species", conn)
print(f"Total animals analyzed: {len(df)}")
df.head()
```

## 2. Environment Distribution (Land, Water, Air)

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='animal_type', palette='viridis')
plt.title('Distribution of Animals by Environment Type')
plt.show()
```

## 3. Top Biological Classes

```python
top_classes = df['class'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_classes.plot(kind='pie', autopct='%1.1f%%', cmap='Pastel1')
plt.title('Top 10 Biological Classes')
plt.ylabel('')
plt.show()
```
